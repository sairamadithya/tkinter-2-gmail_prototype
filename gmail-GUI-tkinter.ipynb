```python
import tkinter
from PIL import Image,ImageTk
import random
import numpy as np
from math import *
global m1,m2,m3,m4,m5,m6,p1,dg,digits,al,alphabets
app=tkinter.Tk()
app.geometry("2700x2700")
app.config(bg='pink')
app.title("G-MAIL")
fn=tkinter.StringVar() 
mn=tkinter.StringVar()
ln=tkinter.StringVar()
age=tkinter.IntVar()
em=tkinter.StringVar()
rep=tkinter.StringVar()
cap=tkinter.StringVar()
def get_mail():
    first=fn.get()
    middle=mn.get()
    last=ln.get()
    aged=age.get()
    m1= str(first)+str(last)+'112'+'@gmail.com'
    m2= str(first)+str(middle)+'234'+'@gmail.com'
    m3= str(last)+str(middle)+str(aged)+'@gmail.com'
    m4= str(last)+str(first)+str(aged)+'@gmail.com'
    m5= str(aged)+str(middle)+str(aged)+'@gmail.com'
    tkinter.Label(app,text=m1,font=("Arial",12,'bold'),fg='black',bg='green').grid(row=21,column=0)
    tkinter.Label(app,text=m2,font=("Arial",12,'bold'),fg='black',bg='green').grid(row=24,column=0)
    tkinter.Label(app,text=m3,font=("Arial",12,'bold'),fg='black',bg='green').grid(row=27,column=0)
    tkinter.Label(app,text=m4,font=("Arial",12,'bold'),fg='black',bg='green').grid(row=30,column=0)
    tkinter.Label(app,text=m5,font=("Arial",12,'bold'),fg='black',bg='green').grid(row=33,column=0)
def mail_verify():
    m6=em.get()
    if(len(m6)>6 and len(m6)<40):
        tkinter.Label(app,text="THIS IS A VALID EMAIL-ID",font=("Arial",14,'bold'),fg='cyan',bg='green').grid(row=43,column=0)
    else:
        tkinter.Label(app,text="THIS IS AN INVALID EMAIL-ID,PLEASE TRY ANOTHER ONE",font=("Arial",14,'bold'),fg='cyan',bg='red').grid(row=43,column=0)

def get_pass():    
    middle=mn.get()
    last=ln.get()
    digits = "0123456789"
    dg = "" 
    for i in range(2) : 
        dg += digits[floor(random.random() * 10)] 
    alphabets="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
    al=""
    for i in range(2) : 
        al += alphabets[floor(random.random() * 51)]     
    p1= str(middle)+str(dg)+str(last)+str(al)
    tkinter.Label(app,text=p1,font=("Arial",12,'bold'),fg='cyan',bg='green').grid(row=50,column=0)
def check_pass():
    middle=mn.get()
    last=ln.get()
    pas2=rep.get()
    digits = "0123456789"
    dg = "" 
    for i in range(2) : 
        dg += digits[floor(random.random() * 10)] 
    alphabets="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
    al=""
    for i in range(2) : 
        al += alphabets[floor(random.random() * 51)]     
    p1= str(middle)+str(dg)+str(last)+str(al)
    if(len(p1)==len(pas2)):
        tkinter.Label(app,text="YOU CAN PROCEED",font=("Arial",12,'bold'),fg='cyan',bg='green').grid(row=58,column=0)
    else:
        tkinter.Label(app,text="PLEASE RE-ENTER THE PASSWORD PROPERLY",font=("Arial",12,'bold'),fg='cyan',bg='red').grid(row=58,column=0)
def gen_img():
    img=Image.open("C:\\Users\\sairam\\Desktop\\captcha.jpeg")
    test = ImageTk.PhotoImage(img)
    label1 = tkinter.Label(image=test)
    label1.image = test
    label1.grid(row=60,column=0)
def prove_human():
    cap1=cap.get()
    if(cap1=='mwxe2'):
        tkinter.Label(app,text="YOU ARE A HUMAN!CONGRAGULATIONS FOR SOLVING THIS COMPLEX TASK",font=("Arial",12,'bold'),fg='cyan',bg='green').grid(row=68,column=0)
    else:
        tkinter.Label(app,text="GET OUT OF HERE STUPID ROBOT",font=("Arial",12,'bold'),fg='cyan',bg='red').grid(row=68,column=0)

tkinter.Label(app,text="ENTER YOUR FIRST NAME",font=("Arial",12),fg='black',bg='pink').grid(row=8,column=0)
tkinter.Entry(app,textvariable=fn).grid(row=8,column=1)
tkinter.Label(app,text="ENTER YOUR MIDDLE NAME",font=("Arial",12),fg='black',bg='pink').grid(row=10,column=0)
tkinter.Entry(app,textvariable=mn).grid(row=10,column=1)
tkinter.Label(app,text="ENTER YOUR LAST NAME",font=("Arial",12),fg='black',bg='pink').grid(row=12,column=0)
tkinter.Entry(app,textvariable=ln).grid(row=12,column=1)
tkinter.Label(app,text="ENTER YOUR AGE",font=("Arial",12),fg='black',bg='pink').grid(row=14,column=0)
tkinter.Entry(app,textvariable=age).grid(row=14,column=1)
tkinter.Label(app,text="CLICK ON THE SUBMIT BUTTON TO GET E-MAIL ID",font=("Arial",12,'bold'),fg='cyan',bg='black').grid(row=17,column=0)
tkinter.Button(app,text="SUBMIT",command=get_mail,bg='yellow',fg='red',font=('bold')).grid(row=19,column=0)

tkinter.Label(app,text="ENTER ONE OF THE EMAIL-ID's",font=("Arial",12),fg='black',bg='pink').grid(row=36,column=0)
tkinter.Entry(app,textvariable=em).grid(row=36,column=1)
tkinter.Label(app,text="CLICK ON THE NEXT BUTTON",font=("Arial",12,'bold'),fg='cyan',bg='black').grid(row=38,column=0)
tkinter.Button(app,text="NEXT",command=mail_verify,bg='yellow',fg='red',font=('bold')).grid(row=40,column=0)

tkinter.Label(app,text="CLICK TO GET A PASSWORD",font=("Arial",12),fg='black',bg='pink').grid(row=46,column=0)
tkinter.Entry(app,textvariable=mn).grid(row=71,column=2)
tkinter.Entry(app,textvariable=ln).grid(row=71,column=2)
tkinter.Button(app,text="CLICK ME",command=get_pass,bg='yellow',fg='red',font=('bold')).grid(row=48,column=0)

tkinter.Label(app,text="RE-ENTER YOUR PASSWORD",font=("Arial",12),fg='black',bg='pink').grid(row=53,column=0)
tkinter.Entry(app,textvariable=rep).grid(row=53,column=1)
tkinter.Entry(app,textvariable=mn).grid(row=71,column=2)
tkinter.Entry(app,textvariable=ln).grid(row=71,column=2)
tkinter.Button(app,text="CLICK TO CHECK",command=check_pass,bg='yellow',fg='red',font=('bold')).grid(row=55,column=0)

tkinter.Button(app,text='CLICK TO GENERATE CAPTCHA',bg='yellow',fg='red',font=('bold'),command=gen_img).grid(row=61,column=0)
tkinter.Label(app,text="ENTER THE CAPTCHA CODE IN THE IMAGE",font=("Arial",12),fg='black',bg='pink').grid(row=63,column=0)
tkinter.Entry(app,textvariable=cap).grid(row=63,column=1)
tkinter.Button(text="PROVE THAT YOU ARE A HUMAN",command=prove_human,bg='yellow',fg='red',font=('bold')).grid(row=65,column=0)

tkinter.Label(app,text="SUCCESSFULLY FINISHED!!",font=("Arial",32,'bold'),fg='cyan',bg='green').grid(row=71,column=2)
tkinter.Label(app,text="THANK YOU",font=("Arial",32,'bold'),fg='green',bg='cyan').grid(row=73,column=2)
app.mainloop()
```
